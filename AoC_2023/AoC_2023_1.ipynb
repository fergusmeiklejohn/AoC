```python
from aocd import get_data

data = get_data(day=1, year=2023)

print(data)
```

# The task is to find the first and last number in each line and combine them
# to form a new number. Then sum all the new numbers.

lines = data.split("\n")

# Find the first and last number in each line and combine them

```python
lines = data.splitlines()

print(lines)
```

```python
total = 0

for line in lines:
    first = None
    last = None
    
    for char in line:
        if char.isdigit():
            if first is None:
                first = char
            last = char
            
    number = int(first + last)
    total += number

print(total)
```

```python
number_map = {word: str(i+1) for i, word in enumerate(["one", "two", "three", "four", "five", "six", "seven", "eight", "nine"])}
print(number_map)
```

Now the task is more complex. We still need to find the first and last number in each line, but that number can be a word or a number. Using the number_map, we can convert the word to a number. So how do we find the first and last number in each line?


```python
total = 0

for line in lines:
    first = None
    last = None
    
    for i in range(len(line)):
        # Check for digit
        if line[i].isdigit():
            if first is None:
                first = line[i]
            last = line[i]
            
        # Check for word numbers
        for word, num in number_map.items():
            if line[i:].startswith(word):            
                if first is None:
                    first = num
                last = num
                
    number = int(first + last)
    total += number

print(total)
```
